```python

# For each non terminal
for name in LAST_SET:

    # For each symbol in LAST(Nt)
    for symbol in LAST_SET[name]:


        # if the symbol is the key of a token, replace it by the name of the token
        if symbol in TOKENS_DICT:
            LAST_SET[name].remove(symbol)
            LAST_SET[name].add(TOKENS_DICT[symbol])


LAST_SET


```
